```python
import math

def ekok_bul(n):
    if n < 1:
        return 0
    
    sonuc = 1
    for i in range(1, n + 1):
        # İki sayının EKOK'u: (a * b) // EBOB(a, b)
        sonuc = (sonuc * i) // math.gcd(sonuc, i)
    
    return sonuc

# Örnek kullanım:
n_degeri = 20
print(f"1'den {n_degeri}'e kadar olan sayıların EKOK'u: {ekok_bul(n_degeri)}")
```
